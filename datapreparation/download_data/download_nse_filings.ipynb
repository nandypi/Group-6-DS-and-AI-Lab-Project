```python
import os
import time
import requests
from datetime import datetime

BASE = "https://www.nseindia.com"
ANNOUNCEMENT_API = BASE + "/api/corporate-announcements"

HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json,text/plain,*/*",
    "Referer": "https://www.nseindia.com/companies-listing/corporate-filings-announcements",
}

session = requests.Session()
session.headers.update(HEADERS)

# Important: hit homepage first so NSE sets cookies
session.get(BASE, timeout=10)

params = {
    "index": "equities",
    "symbol": "INFY",
    "from_date": "01-07-2025",
    "to_date": "01-07-2026",
}

res = session.get(ANNOUNCEMENT_API, params=params, timeout=20)
res.raise_for_status()

announcements = res.json()
print(len(announcements))
print(announcements[0])
```
