```python
# Cell 2: Import Libraries and Setup Paths
import os
from pathlib import Path
from docling.document_converter import DocumentConverter

# Define your Google Drive paths
# Make sure the input directory exactly matches your folder name in MyDrive
INPUT_DIR = Path('/content/drive/MyDrive/infosys_earning_calls_reports')
OUTPUT_DIR = Path('/content/drive/MyDrive/infosys_markdown_results')

# Create the output directory on your Drive if it doesn't exist yet
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print(f"Input Directory configured: {INPUT_DIR}")
print(f"Output Directory configured: {OUTPUT_DIR}")
```

```python
# Cell 3: Batch Convert PDFs to Markdown
print("Initializing Docling Converter (this may take a moment to load models)...")
converter = DocumentConverter()

# Gather all PDF files in the target directory
pdf_files = list(INPUT_DIR.glob("*.pdf"))

if not pdf_files:
    print(f"⚠️ No PDFs found in {INPUT_DIR}. Please double-check the folder name.")
else:
    print(f"Found {len(pdf_files)} PDFs. Starting batch conversion...\n")

    for i, pdf_path in enumerate(pdf_files, 1):
        try:
            print(f"[{i}/{len(pdf_files)}] Processing: {pdf_path.name}")

            # Execute Docling conversion
            result = converter.convert(pdf_path)
            markdown_content = result.document.export_to_markdown()

            # Construct the output filename (swap .pdf for .md)
            output_filename = pdf_path.stem + ".md"
            output_path = OUTPUT_DIR / output_filename

            # Write the markdown payload directly to your Google Drive
            with open(output_path, "w", encoding="utf-8") as f:
                f.write(markdown_content)

            print(f"  ✅ Saved to: {output_filename}")

        except Exception as e:
            print(f"  ❌ ERROR converting {pdf_path.name}: {e}")

print("\n🎉 Batch processing complete! Check your 'infosys_markdown_results' folder in Google Drive.")
```
